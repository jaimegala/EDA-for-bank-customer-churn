```python
gender_exit_counts = df.groupby(['Gender', 'Exited']).size().reset_index(name='counts')
plt.figure(figsize=(8, 6))
sns.barplot(data=gender_exit_counts, x='Gender', y='counts', hue='Exited', palette='plasma')
plt.title('Customer Exits by Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

In this bar plot shows the how many male and female still in this bank and how many exited in this bank
and the result is compare to male and female, Females exited more than males

```python
geography_count = df.groupby(['Geography','Gender','Exited']).size().reset_index(name='counts')
plt.figure(figsize=(8,1))
g =sns.FacetGrid(geography_count, row='Geography', col='Gender',hue = 'Exited')
g.map(sns.barplot,'Exited','counts')
g.add_legend()
plt.tight_layout()
plt.show()
```

above visualization is used for sns facegrid to know how many people stay and leave the bank by geography and gender wise.
The people who are in france exited more compare to germany and spain.Female exited more than the male
